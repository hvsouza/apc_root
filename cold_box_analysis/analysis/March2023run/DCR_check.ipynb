```python
values_h = [29.4, 29.6, 33.2, 31.9, 32, 37.1, 33, 16.7, 14.7, 17.7, 13.7, 33.7, 31.2, 34.6]
values_d = [33.1, 33.3, 37.5, 36.4, 36.7, 42.9, 37.1, 20.1, 16.5, 21.1, 15.6, 37.9, 34.7, 41]
description = ["No PoF(HV ON)",
               "V4 on (HV ON)",
               "V4 and V5 on (HV ON)",
               "V4 and V5 on (HV ON)",
               "V4 and V5 on (HV OFF)",
               "V4 and V5 on (HV OFF)",
               "V4 and V5 on (HV ON again)",
               "V4 and V5 on (HV ON again)",
               "V4 and V5 on (HV ON again)",
               "V4 and V5 on (HV ON)",
               "No PoF (HV ON)",
               "V4 and V5 on (HV ON)",
               "PoF off (HV OFF)",
               "V4 and V5 on (HV OFF)"
               ]

fig, axs = plt.subplots(1, 1, figsize=(15,5))
ind = np.arange(len(values_h))
axs = plt.subplot()
axs.plot(ind, values_d, 'v', color="red", label="Davide Moretti")
axs.plot(ind, values_h, 'o', label='Henrique Souza')
axs.set_xticks(ind)
axs.set_xticklabels(description, rotation = 65, ha='right')
axs.set_ylabel("DCR (Hz/mm$^2$)")
axs.legend(fontsize=12, fancybox=False, edgecolor='black')
axs.axvspan(7, 10, color='grey', alpha=0.3)
axs.text(8,35.5,"Black cover",size=12)
plt.savefig("compare_DCR.png", bbox_inches='tight')
```
