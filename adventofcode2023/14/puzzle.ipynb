```python
import numpy as np
import itertools 
import re
from tqdm import tqdm
f = open("input.dat")
lines = f.readlines()
lines = [line.strip() for line in lines if line.strip()]
f.close()

lines = np.array([[c for c in line] for line in lines])
lines_c_only = np.array([[c if c=="#" else "." for c in line] for line in lines])

def send_nort(lines, lines_c_only):
    lines = lines.transpose()
    lines_c_only = lines_c_only.transpose()
    nrows = len(lines[0])
    lines_ns = [''.join(line) for line in lines]
    lines_c_only_ns = [''.join(line).split('#') for line in lines_c_only]
    lines_b = [''.join(line).split("#") for line in lines]
    lines_b_count = [[c for c in line if c.strip()] for line in lines_b]
    lines_b_count = [[c.count("O") for c in line] for line in lines_b_count]
    lines_b_count_place = [[c.count("O") for c in line] for line in lines_b]
    lines_idxes = [[nrows-i if c!='#' else "-1" for i, c in enumerate(line)] for line in lines_ns]
    lines_idxes = [[list(y) for x, y in itertools.groupby(line, lambda z: z=="-1") if not x] for line in lines_idxes]



    for i, (ln, lconly, lb, lbc, lidx) in enumerate(zip(lines, lines_c_only_ns, lines_b, lines_b_count_place, lines_idxes)):
        # print(ln, lconly, lb, lbc, lidx)
        for j, (n, val) in enumerate(zip(lbc, ln)):
            lines_c_only_ns[i][j] = 'O'*n + lines_c_only_ns[i][j][n:]

    lines_c_only_ns = ['#'.join(line) for line in lines_c_only_ns]
    # for i, (ln, lb, lbc, lidx) in enumerate(zip(lines_c_only_ns, lines, lines_b_count, lines_idxes)):
    #     print(ln, lb, lbc, lidx, len(ln))
    lines = np.array([[c for c in line] for line in lines_c_only_ns]).transpose()
    lines_c_only = np.array([[c if c=="#" else "." for c in line] for line in lines])

    # for line in lines:
    #     print(''.join(line))
    # print()
    return lines, lines_c_only, lines_idxes, lines_b_count

def send_west(lines, lines_c_only):
    nrows = len(lines[0])
    lines_ns = [''.join(line) for line in lines]
    lines_c_only_ns = [''.join(line).split('#') for line in lines_c_only]
    lines_b = [''.join(line).split("#") for line in lines]
    lines_b = [[c for c in line] for line in lines_b]
    lines_b_count = [[c.count("O") for c in line] for line in lines_b]
    lines_idxes = [[nrows-i if c!='#' else "-1" for i, c in enumerate(line)] for line in lines_ns]
    lines_idxes = [[list(y) for x, y in itertools.groupby(line, lambda z: z=="-1") if not x] for line in lines_idxes]

    for i, (ln, lconly, lb, lbc) in enumerate(zip(lines, lines_c_only_ns, lines_b, lines_b_count)):
        # print(ln, lconly, lb, lbc, lidx)
        for j, (n, val) in enumerate(zip(lbc, ln)):
            lines_c_only_ns[i][j] = 'O'*n + lines_c_only_ns[i][j][n:]

    lines_c_only_ns = ['#'.join(line) for line in lines_c_only_ns]
    lines = np.array([[c for c in line] for line in lines_c_only_ns])
    lines_c_only = np.array([[c if c=="#" else "." for c in line] for line in lines])

    # for line in lines:
    #     print(''.join(line))
    # print()
    return lines, lines_c_only#, lines_idxes, lines_b_count
def send_south(lines, lines_c_only):
    lines = lines.transpose()
    lines_c_only = lines_c_only.transpose()
    nrows = len(lines[0])
    lines_ns = [''.join(line) for line in lines]
    lines_c_only_ns = [''.join(line).split('#') for line in lines_c_only]
    lines_b = [''.join(line).split("#") for line in lines]
    lines_b = [[c for c in line] for line in lines_b]
    lines_b_count = [[c.count("O") for c in line] for line in lines_b]

    for i, (ln, lconly, lb, lbc) in enumerate(zip(lines, lines_c_only_ns, lines_b, lines_b_count)):
        for j, (n, val) in enumerate(zip(lbc, ln)):
            tmplen = len(lines_c_only_ns[i][j])
            lines_c_only_ns[i][j] = lines_c_only_ns[i][j][:tmplen-n]+'O'*n 

    lines_c_only_ns = ['#'.join(line) for line in lines_c_only_ns]
    lines = np.array([[c for c in line] for line in lines_c_only_ns]).transpose()
    lines_c_only = np.array([[c if c=="#" else "." for c in line] for line in lines])

    # for line in lines:
    #     print(''.join(line))
    # print()
    return lines, lines_c_only

def send_east(lines, lines_c_only):
    nrows = len(lines[0])
    lines_ns = [''.join(line) for line in lines]
    lines_c_only_ns = [''.join(line).split('#') for line in lines_c_only]
    lines_b = [''.join(line).split("#") for line in lines]
    lines_b_count = [[c for c in line if c.strip()] for line in lines_b]
    lines_b_count = [[c.count("O") for c in line] for line in lines_b_count]
    lines_b_count_place = [[c.count("O") for c in line] for line in lines_b]
    lines_idxes = [[nrows-i if c!='#' else "-1" for i, c in enumerate(line)] for line in lines_ns]
    lines_idxes = [[list(y) for x, y in itertools.groupby(line, lambda z: z=="-1") if not x] for line in lines_idxes]

    for i, (ln, lconly, lb, lbc) in enumerate(zip(lines, lines_c_only_ns, lines_b, lines_b_count_place)):
        # print(ln, lconly, lb, lbc, lidx)
        for j, (n, val) in enumerate(zip(lbc, ln)):
            tmplen = len(lines_c_only_ns[i][j])
            lines_c_only_ns[i][j] = lines_c_only_ns[i][j][:tmplen-n]+'O'*n 

    lines_c_only_ns = ['#'.join(line) for line in lines_c_only_ns]
    lines = np.array([[c for c in line] for line in lines_c_only_ns])
    lines_c_only = np.array([[c if c=="#" else "." for c in line] for line in lines])

    return lines, lines_c_only, lines_idxes, lines_b_count


lines_before = []
id_repeat = 0
total_cicles = 1000000000
for cicle in tqdm(range(total_cicles)):
    # north
    lines, lines_c_only, lines_idxes, lines_b_count = send_nort(lines, lines_c_only)
    lines, lines_c_only = send_west(lines, lines_c_only)
    lines, lines_c_only = send_south(lines, lines_c_only)
    lines, lines_c_only, lines_idxes, lines_b_count = send_east(lines, lines_c_only)

    # for i, line in enumerate(lines):
    #     print(''.join(line), len(line)-i)
    # print()
    if id_repeat == 0:
        for idx, lines_b in enumerate(lines_before):
            # print(lines_b, lines)
            if np.all(lines == lines_b):
                id_repeat = idx
                repeat_each = cicle-idx
                count = 0
                this_extra = (total_cicles-(id_repeat+1))%repeat_each
                print(idx, cicle, repeat_each, this_extra)
        lines_before.append([[c for c in line] for line in lines])
    else:
        if count == this_extra-1:
            break
        count+=1

# lines, lines_c_only, lines_idxes, lines_b_count = send_nort(lines, lines_c_only)

retline = 0
for i, line in enumerate(lines):
    line = ''.join(line)
    retline += line.count("O")*(len(line)-i)

print(retline)
```
