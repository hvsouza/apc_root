```python

import numpy as np
f = open("input.dat")
lines = f.readlines()
f.close()
lines = [line.strip() for line in lines if line.strip()]

idx_add_line = []
for i, line in enumerate(lines):
     if not "#" in line:
        idx_add_line.append(i)


linesv = np.array([[c for c in line] for line in lines])
idx_add_col = []
for i, line in enumerate(linesv.transpose()):
    if not "#" in line:
        idx_add_col.append(i)


for line in lines:
    print(line)

galaxies = []
gnumber = 0
for i, line in enumerate(lines):
    galaxies_l = [j for j, c in enumerate(line) if c == "#"]
    if len(galaxies_l)>0:
        for j in galaxies_l:
            galaxies.append([i,j, gnumber])
            gnumber+=1

pairs = []
for i, g in enumerate(galaxies):
    if i == len(galaxies)-1: break
    for g2 in galaxies[i+1:]:
        pairs.append([g, g2])

def get_true_idx(g):
    factor = 1000000
    truei, truej = g[0], g[1]
    i, j = g[0], g[1]

    for idx_line in reversed(idx_add_line):
        if idx_line < i:
            truei+=(factor-1)
        
    for idx_col in reversed(idx_add_col):
        if idx_col < j:
            truej+=(factor-1)

    return [truei, truej]
    
distance = 0
for [g1, g2] in pairs:
    g1 = get_true_idx(g1)
    g2 = get_true_idx(g2)
    distance_x = abs(g2[0]-g1[0])
    distance_y = abs(g2[1]-g1[1])
    # if distance_y>0: distance_y -= 1
    distance += distance_x + distance_y
    # if g1[2] == 4 and g2[2] == 8:
    #     print(g1,g2, distance, distance_x, distance_y)
print(distance)

    
```
