```python
import numpy as np
f = open("input.dat","r")
lines = f.readlines()
lines = [line.strip() for line in lines]
f.close()


score = 0
ncopies = [1]*len(lines)
for line in lines:
    card_id = int(line.split(":")[0].split()[1])
    card_idx = card_id - 1
    tmp = line.split(":")[1]
    winner = tmp.split('|')[0].split()
    mynumb = tmp.split('|')[1].split()
    winner = np.array([int(v) for v in winner])
    mynumb = np.array([int(v) for v in mynumb])
    matches = [1 for v in mynumb if np.any(v==winner)]
    nmatches = len(matches)
    if nmatches>=1:
        score += 2**(nmatches-1)
        card_copies = ncopies[card_idx]
        for i in range(card_idx+1, card_idx+1+nmatches):
            ncopies[i]+=1*card_copies
        # print(f"id:{card_id}, total:{card_copies},  nmatches:{nmatches}, ncopies:{ncopies}")

        
print(f'Part one: {score}')
print(f'Part two: {sum(ncopies)}')
```
